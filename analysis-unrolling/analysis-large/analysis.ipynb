```python
import pandas as pd
import seaborn as sns
```

```python
df_inc = pd.read_csv("inc.csv")
df_movl = pd.read_csv("movl.csv")
df_nop = pd.read_csv("nop.csv")
```

```python
df_inc['Count'] = df_inc.groupby('Measurements')['Measurements'].transform('size')
df_movl['Count'] = df_movl.groupby('Measurements')['Measurements'].transform('size')
df_nop['Count'] = df_nop.groupby('Measurements')['Measurements'].transform('size')
```

```python
df_inc = df_inc.drop_duplicates()
df_movl = df_movl.drop_duplicates()
df_nop = df_nop.drop_duplicates()
```

## Analysis of inc ##

```python
sns.lineplot(x="Measurements",y="Count",data=df_inc)
```

```python
df_inc = df_inc[(df_inc[['Measurements']] != 0).all(axis=1)]
df_inc = df_inc[(df_inc[['Measurements']] < 7000).all(axis=1)]
df_inc = df_inc[(df_inc[['Measurements']] > 5000).all(axis=1)]
```

```python
sns.lineplot(x="Measurements",y="Count",data=df_inc)
```

## Analysis of movl ##

```python
sns.lineplot(x="Measurements",y="Count",data=df_movl)
```

```python
df_movl = df_movl[(df_movl[['Measurements']] != 0).all(axis=1)]
df_movl = df_movl[(df_movl[['Measurements']] < 8000).all(axis=1)]
df_movl = df_movl[(df_movl[['Measurements']] > 6000).all(axis=1)]
```

```python
sns.lineplot(x="Measurements",y="Count",data=df_movl)
```

## Analysis of nop ##

```python
sns.lineplot(x="Measurements",y="Count",data=df_nop)
```

```python
df_nop = df_nop[(df_nop[['Measurements']] != 0).all(axis=1)]
df_nop = df_nop[(df_nop[['Measurements']] < 8000).all(axis=1)]
df_nop = df_nop[(df_nop[['Measurements']] > 6000).all(axis=1)]
```

```python
sns.lineplot(x="Measurements",y="Count",data=df_nop)
```
