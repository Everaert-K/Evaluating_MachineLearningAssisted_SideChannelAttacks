```python
import pandas as pd
import seaborn as sns
```

```python
df_fscale = pd.read_csv("fscale.csv")
df_inc = pd.read_csv("inc.csv")
df_movl = pd.read_csv("movl.csv")
df_nop = pd.read_csv("nop.csv")
df_rdrand = pd.read_csv("rdrand.csv")
df_xor = pd.read_csv("xor.csv")
```

```python
df_fscale['Count'] = df_fscale.groupby('Measurements')['Measurements'].transform('size')
df_inc['Count'] = df_inc.groupby('Measurements')['Measurements'].transform('size')
df_movl['Count'] = df_movl.groupby('Measurements')['Measurements'].transform('size')
df_nop['Count'] = df_nop.groupby('Measurements')['Measurements'].transform('size')
df_rdrand['Count'] = df_rdrand.groupby('Measurements')['Measurements'].transform('size')
df_xor['Count'] = df_xor.groupby('Measurements')['Measurements'].transform('size')
```

```python
df_fscale = df_fscale.drop_duplicates()
df_inc = df_inc.drop_duplicates()
df_movl = df_movl.drop_duplicates()
df_nop = df_nop.drop_duplicates()
df_rdrand = df_rdrand.drop_duplicates()
df_xor = df_xor.drop_duplicates()
```

## Analysis of fscale ##

```python
sns.lineplot(x="Measurements",y="Count",data=df_fscale)
```

```python
df_fscale = df_fscale[(df_fscale[['Measurements']] != 0).all(axis=1)]
df_fscale = df_fscale[(df_fscale[['Measurements']] < 580000).all(axis=1)]
df_fscale = df_fscale[(df_fscale[['Measurements']] > 540000).all(axis=1)]
```
